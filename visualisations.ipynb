```python
keys = ["_0_", "_5_", "_25_", "_50_", "_100_", "_200_", "_500_"]
f1s = {k:0 for k in keys}
print(f1s)

for k,v in stats1[3].items():
    for key in keys:
        if key in k:
            f1s[key] += v
            
for k,v in stats2[3].items():
    for key in keys:
        if key in k:
            f1s[key] += v

print(f1s)
print(np.array(list(f1s.values())) / 5)

GAT_data = np.array(list(f1s.values())) / 5
```

```python
x = [0, 5, 25, 50, 100, 200, 500]

 # data from manual runs 
GCN_data = [0.706, 0.701, 0.714, 0.716, 0.718, 0.713, 0.693] # to update




fig, ax = plt.subplots(1,1, figsize=(8,6))


ax.plot(x, GAT_data, marker= "o", linestyle="--", label = "rGAT", linewidth=3, ms=10)
ax.plot(x, GCN_data, marker= "o", linestyle="--", label = "rGCN", linewidth=3, ms=10)

ax.set_xlabel("Number of random features", fontsize=20)
ax.set_ylabel("F1 score", fontsize=20)
ax.set_ylim([0.6, 1])
ax.set_title("F1 Score against # random features", fontsize=20)
ax.set_xscale("linear")
ax.tick_params(axis='both', which='major', labelsize=14)
#ax.tick_params(axis='both', which='major', labelsize=10)
ax.legend(fontsize=20)

    
    
plt.show()
```
